```python
import time

name = input("Welcome! Please enter your name: ")
print("Welcome,", name + ", to this hypothetical adventure game.")

points = 0

while True:
    print("\nYou find yourself at the end of a dirt road, surrounded by nature's beauty.")
    time.sleep(1)
    direction = input("You can go left into the dense jungle or right towards a nearby village. Choose wisely: ").lower()

    if direction == 'left':
        print("\nYou venture into the dense jungle, the air thick with humidity and the sounds of wildlife.")
        time.sleep(1)
        exploration_decision = input("Do you want to push deeper into the jungle or turn back? (Type 'explore' or 'turn back'): ").lower()
        
        if exploration_decision == 'explore':
            points += 10
            print("\nYou press on, determined to uncover the secrets hidden within the jungle.")
            time.sleep(1)
            print("Suddenly, you stumble upon an ancient temple hidden among the foliage.")
            time.sleep(1)
            print("You cautiously enter the temple, discovering treasures beyond your wildest dreams!")
            points += 30
        elif exploration_decision == 'turn back':
            print("\nFeeling uncertain, you decide to turn back, fearing the unknown.")
            points -= 5
        else:
            print("\nInvalid choice. Your indecision leaves you feeling uneasy.")
            points -= 5
            
    elif direction == 'right':
        print("\nYou head towards the nearby village, eager to seek the company of fellow travelers.")
        time.sleep(1)
        village_decision = input("As you approach, you notice festivities in full swing. Will you join the celebration or continue on? (Type 'join' or 'continue'): ").lower()
        
        if village_decision == 'join':
            points += 20
            print("\nYou join the villagers in their festivities, dancing and feasting into the night.")
            time.sleep(1)
            print("The villagers welcome you with open arms, grateful for your company.")
        elif village_decision == 'continue':
            print("\nYou choose to continue on your journey, resisting the allure of the village's festivities.")
            time.sleep(1)
            print("Your path leads you towards new adventures, leaving the village behind.")
        else:
            print("\nInvalid choice. Your hesitation causes you to miss out on the village's hospitality.")
            points -= 5
            
    else:
        print("\nNot a valid option. You stand indecisively, unsure of which path to take.")
        points -= 5
    
    print("\nCurrent Points:", points)
    play_again = input("Do you want to continue your adventure? (Type 'yes' or 'no'): ").lower()
    if play_again != 'yes':
        break

print("\nThanks for playing, {}. Your final score is: {} points!".format(name, points))
```
